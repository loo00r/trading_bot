```python
def feature_engineering(df):
    #    
    
    #   
    df['hl_range'] = df['high'] - df['low']
    df['oc_range'] = df['open'] - df['close']

    df['RSI_diff'] = df['RSI_8MIN'] - df['RSI']
    
    # Bollinger Bands
    # df['bb_width'] = (df['BBU_14_2.0'] - df['BBL_14_2.0']) / df['BBM_14_2.0']
    # df['bb_position'] = (df['close'] - df['BBL_14_2.0']) / (df['BBU_14_2.0'] - df['BBL_14_2.0'])
    
    #  
    # df['log_return'] = np.log(df['close'] / df['close'].shift(1))
    # df['log_volume_change'] = np.log(df['volume'] / df['volume'].shift(1))
        
    df['close_vwap_ratio'] = df['close'] / df['VWAP'] - 1
    
    # #  VWAP
    # df['vwap_change'] = df['VWAP'].pct_change()
                   
    #  SMA  EMA
    df['SMA_10'] = df['close'].rolling(window=7).mean()
    df['EMA_5'] = df['close'].ewm(span=4, adjust=False).mean()
    
    #      SMA  EMA
    df['close_between_ma'] = (df['close'] - df['SMA_10']) / (df['EMA_5'] - df['SMA_10'])
    
    #   EMA  SMA

            
    #  SMA  EMA
    # df['sma_slope'] = (df['SMA_10'] - df['SMA_10'].shift(1)) / df['SMA_10'].shift(1)
    # df['ema_slope'] = (df['EMA_5'] - df['EMA_5'].shift(1)) / df['EMA_5'].shift(1)

    # Ichimoku differences
    df['ichimoku_tenkan_kijun_diff'] = df['tenkan_sen'] - df['kijun_sen']
    df['ichimoku_close_kijun_diff'] = df['close'] - df['kijun_sen']
    df['ichimoku_span_a_b_diff'] = df['span_a'] - df['span_b']

     
    df = df.drop(['EMA_5', 'SMA_10', 'high', 'open', 'low', 'close', 'tenkan_sen', 'kijun_sen', 'span_a', 'span_b', 'chikou_span', 'RSI', 'MACD_12_26_9', 'MACDs_12_26_9'], axis=1)
    



    return df
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense, Dropout, Attention, Input, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Dataset setup with 12 features
seq_length = 50
pred_length = 5
n_features = 12

# Function to create sequences
def create_sequences(data, seq_length, pred_length):
    X, y = [], []
    for i in range(len(data) - seq_length - pred_length + 1):
        X.append(data[i:i+seq_length])  # Training sequence
        y.append(data[i+seq_length:i+seq_length+pred_length, 5])  # Predict "oc_range" (index 5)
    return np.array(X), np.array(y)

# Custom loss function (MSE)
def custom_loss(y_true, y_pred):
    mse = tf.reduce_mean(tf.square(y_true - y_pred), axis=-1)
    return mse

# Define LSTM model with Attention layer
def build_lstm_attention_model(seq_length, n_features, pred_length, l2_value=0.0005, dropout_rate=0.2):
    input_seq = Input(shape=(seq_length, n_features))
    
    # First LSTM layer with Dropout
    x = LSTM(50, return_sequences=True, kernel_regularizer=l2(l2_value))(input_seq)
    x = Dropout(dropout_rate)(x)
    
    # Second LSTM layer with Dropout
    x = LSTM(50, return_sequences=True, kernel_regularizer=l2(l2_value))(x)
    x = Dropout(dropout_rate)(x)
    
    # Attention layer with reduction
    attention = Attention()([x, x])  # [batch_size, seq_length, seq_length]
    attention_summary = tf.reduce_mean(attention, axis=1)  # [batch_size, seq_length]
    
    # Output layer for predictions
    output = Dense(pred_length)(attention_summary)  # [batch_size, pred_length]
    
    # Define and compile the model
    model = Model(inputs=input_seq, outputs=output)
    model.compile(optimizer='adam', loss=custom_loss)
    
    return model


# Walk-forward validation parameters
test_size = int(len(scaled_data) * 0.2)  # 20% test set
initial_train_size = len(scaled_data) - test_size
step_size = int(test_size / 5)  # Step size per block

for i in range(0, test_size, step_size):
    train_data = scaled_data[:initial_train_size + i]
    test_data = scaled_data[initial_train_size + i:initial_train_size + i + step_size]
    
    # Prepare sequences
    X_train, y_train = create_sequences(train_data, seq_length, pred_length)
    X_test, y_test = create_sequences(test_data, seq_length, pred_length)
    
    # Build and train the model
    model = build_lstm_attention_model(seq_length=seq_length, n_features=n_features, pred_length=pred_length)
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)
    
    # Evaluate on the test set
    loss = model.evaluate(X_test, y_test, verbose=0)  # Direct float output for single loss
    print(f"Test Loss for block {i // step_size + 1}: {loss}")

# Visualize Attention Weights for Feature Importance
# Update model definition to return attention weights for visualization
sample_seq = X_train[0:1]  # Single example sequence
attention_model = Model(inputs=model.input, outputs=[model.output, model.get_layer('attention_4').output])
predictions, attention_weights = attention_model.predict(sample_seq)

# Plotting the attention weights
plt.figure(figsize=(10, 6))
plt.imshow(attention_weights[0], cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Attention Weights for Each Timestep and Feature")
plt.xlabel("Features")
plt.ylabel("Timesteps")
plt.show()

```
